# KDA rate (trash)

```python
# Total
total_kills_100 = df['100TOP_kills'] + df['100JUN_kills'] + df['100MID_kills'] + df['100ARC_kills'] + df['100UTIL_kills']
total_kills_200 = df['200TOP_kills'] + df['200JUN_kills'] + df['200MID_kills'] + df['200ARC_kills'] + df['200UTIL_kills']

total_assists_100 = df['100TOP_assists'] + df['100JUN_assists'] + df['100MID_assists'] + df['100ARC_assists'] + df['100UTIL_assists']
total_assists_200 = df['200TOP_assists'] + df['200JUN_assists'] + df['200MID_assists'] + df['200ARC_assists'] + df['200UTIL_assists']

total_deaths_100 = df['100TOP_deaths'] + df['100JUN_deaths'] + df['100MID_deaths'] + df['100ARC_deaths'] + df['100UTIL_deaths']
total_deaths_200 = df['200TOP_deaths'] + df['200JUN_deaths'] + df['200MID_deaths'] + df['200ARC_deaths'] + df['200UTIL_deaths']

# Prevent div0
total_kills_100 = total_kills_100 + 1e-3
total_assists_100 = total_assists_100 + 1e-3
total_deaths_100 = total_deaths_100 + 1e-3

total_kills_200 = total_kills_200 + 1e-3
total_assists_200 = total_assists_200 + 1e-3
total_deaths_200 = total_deaths_200 + 1e-3
```

```python
dict_rate = {
    'teamId': df['teamId'], 
    'kills': df['kills'],
    'assists': df['assists'],
    'deaths': df['deaths'],
    # '100kills': total_kills_100,
    # '100assists': total_assists_100,
    # '100deaths': total_deaths_100,
    # '200kills': total_kills_200,
    # '200assists': total_assists_200,
    # '200deaths': total_deaths_200,
    'win': df['win'],
    
    }

df_rate = pd.DataFrame(dict_rate)
df_rate

# 내일할일: 팀id 100 200에 따라 합산된 KDA를 이용하여 비율을 구하고 상관관계를 알아볼것.    
```

```python
# Divide teamId
is_team100 = df_rate['teamId'] == 100
is_team200 = df_rate['teamId'] == 200

subset_100 = df_rate[is_team100]
subset_200 = df_rate[is_team200]
```

```python
subset_100
```

```python
subset_200
```

```python
subset_100['killRate'] = subset_100['kills'] / total_kills_100
subset_100['assistRate'] = subset_100['assists'] / total_assists_100
subset_100['deathRate'] = subset_100['deaths'] / total_deaths_100

subset_200['killRate'] = subset_200['kills'] / total_kills_200
subset_200['assistRate'] = subset_200['assists'] / total_assists_200
subset_200['deathRate'] = subset_200['deaths'] / total_deaths_200
```

```python
subset_100
```

```python
subset_200
```

```python
df_rate = pd.concat([subset_100, subset_200])
df_rate = df_rate.sort_index()

a = df_rate.iloc[:, :4]
b = df_rate.iloc[:, 5:]
c = df_rate.iloc[:, 4]

df_rate = pd.concat([a, b, c], axis=1)
df_rate.drop(['teamId', 'kills', 'assists', 'deaths'], axis=1, inplace=True)
df_rate
```

# Gold correlation (goldDiff, goldEarned)

## * goldDiff = myteamGold - awayteamGold

```python
df_gold = {
    'teamId': df['teamId'],
    'gameDuration': df['gameDuration'],
    'goldEarned': df['goldEarned'],
    '100teamGold': df['100TOP_goldEarned'] + df['100JUN_goldEarned'] + df['100MID_goldEarned'] + df['100ARC_goldEarned'] + df['100UTIL_goldEarned'],
    '200teamGold': df['200TOP_goldEarned'] + df['200JUN_goldEarned'] + df['200MID_goldEarned'] + df['200ARC_goldEarned'] + df['200UTIL_goldEarned'],
    'win': df['win']

}

df_gold = pd.DataFrame(df_gold)
df_gold
```

```python
is_team100 = df_gold['teamId'] == 100
is_team200 = df_gold['teamId'] == 200

subset_100 = df_gold[is_team100]
subset_200 = df_gold[is_team200]
```

```python
subset_100
```

```python
subset_200
```

```python
subset_100['goldRate'] = subset_100['goldEarned'] / subset_100['100teamGold']
subset_100['goldDiff'] = subset_100['100teamGold'] - subset_100['200teamGold']

subset_200['goldRate'] = subset_200['goldEarned'] / subset_200['200teamGold']
subset_200['goldDiff'] = subset_200['200teamGold'] - subset_200['100teamGold']
```

```python
a = subset_100.iloc[:, :5]
b = subset_100.iloc[:, 6:]
c = subset_100.iloc[:, 5]
subset_100 = pd.concat([a, b, c], axis=1)

a = subset_200.iloc[:, :5]
b = subset_200.iloc[:, 6:]
c = subset_200.iloc[:, 5]
subset_200 = pd.concat([a, b, c], axis=1)
```

```python
df_gold = pd.concat([subset_100, subset_200]).sort_index()
df_gold
```
